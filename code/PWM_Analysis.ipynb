### Verifying against PsyTrack

```python
checkP = np.load(f'../data/P_1_state_PWM.npy')
checkLl = np.load(f'../data/Ll_1_state_PWM.npy')
checkW = np.load(f'../data/W_1_state_PWM.npy')

fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True, sharey=False, figsize=(12,10))
def plotting_weights_PWM(w, sessInd, axes, title=''):
    K = w.shape[1]
    sess = len(sessInd)-1
    axes[0].plot(range(1,sess+1),w[sessInd[:-1],0,1,0],color='indianred',linewidth=5,label='state 0 stim A')
    axes[0].plot(range(1,sess+1),w[sessInd[:-1],0,0,0],color='gold',linewidth=5, label='state 0 bias')
    axes[0].plot(range(1,sess+1),w[sessInd[:-1],0,2,0],color='cornflowerblue',linewidth=5, label='state 0 stim B')

    axes[0].set_title(title)
    #plt.xticks(range(1,sess+1))
    axes[0].set_ylabel("weights")
    axes[0].set_xlabel('session')
    #plt.legend(fontsize='xx-small')

    axes[1].plot(range(1,sess+1),w[sessInd[:-1],0,4,0],color='thistle', linewidth=5, label='state 0 prev_choice')
    axes[1].plot(range(1,sess+1),w[sessInd[:-1],0,5,0],color='mediumturquoise',linewidth=5, label='state 0 prev_correct')
    axes[1].plot(range(1,sess+1),w[sessInd[:-1],0,3,0],color='lightgreen',linewidth=5, label='state 0 prev_s')

    axes[1].set_ylim(-0.2,2)
    #plt.xticks(range(1,sess+1))
    axes[1].set_ylabel("weights")
    axes[1].set_xlabel('session')

    if (K >= 2):
        axes[0].plot(range(1,sess+1),w[sessInd[:-1],1,1,0],color='darkred',marker='o',label='state 0 stim A')

    axes[0].legend()
    axes[1].legend()


plotting_weights_PWM(w=checkW, sessInd=sessInd, axes=axes)
```
